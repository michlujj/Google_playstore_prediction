```python
# to remove unimportant variables before predictive modelling
df = df.drop(['App','Genres'], axis=1)
```

```python
# to define X and y
X = df.drop('Rating', axis=1)
```

```python
# to convert categorical variables to dummy variables
X = pd.get_dummies(X, drop_first=True)
```

```python
X
```

```python
y = df['Rating']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
# testing data = 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# to initialize the standard scaler
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
dt = DecisionTreeRegressor()
```

```python
# to fit the decision tree modelinto the training data
dt.fit(scaled_X_train, y_train)
```

```python
# Decision tree train model has a perfect score of 1, but a negative testing score
# the training model is overfitted and it performs poorly on the testing set, resulting incorrect predictions
print(f"Decision tree_training set score: {dt.score(scaled_X_train, y_train):.2f}")
print(f"Decision tree_testing set score: {dt.score(scaled_X_test, y_test):.2F}") 
```

```python
# to do cross validation to determine which alpha in Ridge regression to use
from sklearn.linear_model import RidgeCV
```

```python
# perform cross validation to determine which alpha value to use in Ridge regression
# alpha = 0.1, 1.0, 10.0 are default values in Ridge CV model
ridge_cv_model = RidgeCV(alphas=(0.1, 1.0, 10.0))
```

```python
# to use the training set for the purpose of Hyper parameter tuning for Ridge regression model
ridge_cv_model.fit(scaled_X_train, y_train)
```

```python
# to see which alpha value perform the best, alpha= 1.0 performs the best
ridge_cv_model.alpha_
```

```python
# neg_mean_absolute_error, the higher the value the better, to re-run model with the scoring metric
ridge_cv_model = RidgeCV(alphas=(0.1, 1.0, 10.0), scoring='neg_mean_absolute_error')
```

```python
ridge_cv_model.fit(scaled_X_train, y_train)
```

```python
# to see which alpha value perform the best after using the scoring metric, alpha= 1.0 performs the best
ridge_cv_model.alpha_
```

```python
ridge_cv_model.coef_[:10]
```

```python
# to import LASSO CV for hyperparameter tuning to prevent overfitting
from sklearn.linear_model import LassoCV
```

```python
# default values: eps= 0.01, n_alphas=100, to increase no. of max iterations
lasso_cv_model = LassoCV(eps=0.001, n_alphas=100, cv=5, max_iter=1000000)
```

```python
# to fit the Lasso Cross validation on training data
lasso_cv_model.fit(scaled_X_train, y_train)
```

```python
# to check the best performing alpha value in Lasso Regression
lasso_cv_model.alpha_
```

```python
lasso_cv_model.coef_[:10]
```

```python
# ElasticNet is a combination of L1 (Ridge) and L2 (LASSO) regularization
from sklearn.linear_model import ElasticNet
```

```python
# to initialise an instance of ElasticNet
el = ElasticNet(max_iter=1000000)
```

```python
# elastic net model has 2 parameters: alpha and L1 ratio
# to create a dictionary parameter grid values for ElasticNet
param_grid = {'alpha':[0.1,1,5,10,50,100],
              'l1_ratio':[.1, .5, .7, .9, .95, .99, 1]}
```

```python
# to use GridSearchCV to run grid search for the best parameter for ElasticNet
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=el,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5,
                          verbose=1)
```

```python
grid_model.fit(scaled_X_train,y_train)
```

```python
# to display best combination parameter for Elastic Net GridSearch CV
grid_model.best_params_
```

```python
# to use regularization Elastic Net to predict on X test and training set on Decision tree
dt_pred = grid_model.predict(scaled_X_test)
dt_pred[:10]
```

```python
dt_train_pred = grid_model.predict(scaled_X_train)
dt_train_pred[:10]
```

```python
# MSE for Decision tree test model
mean_squared_error(y_test, dt_pred)
```

```python
# MSE for Decision tree train model
mean_squared_error(y_train, dt_train_pred)
```

```python
print('Decision Tree_test_RMSE:', np.sqrt(mean_squared_error(y_test, dt_pred)))  # Decision tree RMSE test model = 0.56
print('Decision Tree_train_RMSE:', np.sqrt(mean_squared_error(y_train, dt_train_pred)))
```
